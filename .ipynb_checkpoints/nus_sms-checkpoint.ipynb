## Conduct Text Preprocessing

The goal of text preprocessing is to clean the text data, removing irrelevant parts and preparing it for analysis. We will perform the following preprocessing techniques:

1. **Tokenization**: Splitting the text into individual words (tokens).
2. **Noise Removal**: Removing unwanted characters such as:
   - Punctuation
   - Digits
   - Symbols
   - HTML tags
   - URLs
3. **Stopword Removal**: Removing common words that don't contribute to the analysis (e.g., "the", "is").
4. **Text Normalization**:
   - **Lowercasing**: Converting all text to lowercase to maintain consistency.
   - **Stemming/Lemmatization**: Reducing words to their base forms to avoid duplication in analysis.

Each of these steps will help distill the important parts of the text and improve the effectiveness of our NLP models and techniques.


### Tokenization and Noise Removal

In this step, we will preprocess the SMS messages by converting the text to lowercase, removing URLs, HTML tags, punctuation, digits, and special symbols. After that, we will tokenize the cleaned text to prepare it for further analysis.


```python
import re
import nltk
from nltk.tokenize import word_tokenize

# Download stopwords if not already installed
nltk.download('punkt')
nltk.download('stopwords')

# Function for tokenization and noise removal
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs, HTML tags, punctuation, digits, and special symbols
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation and symbols
    
    # Tokenize the cleaned text
    tokens = word_tokenize(text)
    
    return tokens

# Apply the preprocessing function to the 'Message' column
df['tokens'] = df['Message'].apply(lambda x: preprocess_text(str(x)))

# Display the first few rows to check the tokens
df[['Message', 'tokens']].head()
```


In this step, we have:
- **Tokenized** the messages, splitting each message into individual words (tokens).
- **Removed noise**, including:
   - URLs
   - HTML tags
   - Punctuation
   - Digits
   - Special symbols

For example:
- The message "Go until jurong point, crazy.. Available only..." has been tokenized to `[go, until, jurong, point, crazy, available, only]`.
- The message "Meet after lunch la..." has been tokenized to `[meet, after, lunch, la]`.

The result is a clean, tokenized version of the messages, retaining only meaningful words, and is now ready for further preprocessing steps.


### Stopword Removal

In this step, we will remove common **stopwords** such as "the", "is", and "and", which do not contribute significant meaning to our analysis. By removing these words, we focus on the most important terms in the text that carry valuable information.


```python
from nltk.corpus import stopwords

# Load the list of stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords from tokens
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

# Apply the stopword removal function
df['tokens_no_stopwords'] = df['tokens'].apply(remove_stopwords)

# Display the first few rows after stopword removal
df[['tokens', 'tokens_no_stopwords']].head()
```

### Interpretation of Stopword Removal

After removing stopwords from the tokenized text, we can observe that common words, such as "i", "until", and "only", have been successfully removed. The remaining tokens are more meaningful for analysis.

Examples:
- The message "Go until jurong point, crazy.. Available only..." has been reduced to `[go, jurong, point, crazy, available]`, where stopwords like "until" and "only" have been removed.
- The message "Meet after lunch la..." has been reduced to `[meet, lunch, la]`, removing the word "after".

Stopword removal helps focus on the more significant words in the messages, making the data cleaner for further analysis.


### Text Normalization: Stemming and Lemmatization

In this step, we will further normalize the tokenized text by:
- **Stemming**: Reducing words to their root forms (e.g., "running" becomes "run").
- **Lemmatization**: Converting words to their base dictionary forms (e.g., "better" becomes "good").

This process will help reduce variations in word forms, improving consistency in our text data for subsequent analysis.


```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to apply stemming and lemmatization
def normalize_text(tokens):
    # Apply stemming
    stemmed = [stemmer.stem(word) for word in tokens]
    
    # Apply lemmatization
    lemmatized = [lemmatizer.lemmatize(word) for word in stemmed]
    
    return lemmatized

# Apply normalization to the tokens without stopwords
df['normalized_tokens'] = df['tokens_no_stopwords'].apply(normalize_text)

# Display the first few rows after normalization
df[['tokens_no_stopwords', 'normalized_tokens']].head()
```

### Interpretation of Stemming and Lemmatization

After applying stemming and lemmatization, the tokens have been reduced to their root forms. This helps in normalizing variations of words, which simplifies the text for further analysis.

Examples:
- The word "crazy" has been stemmed to "crazi" and "available" to "avail".
- In the message "dunno, lets, go, learn, pilates", the word "lets" has been lemmatized to "let", and "pilates" to "pilat".
- The word "weekdays" has been normalized to "weekday", removing plural forms.

This normalization process reduces the dimensionality of the text data, making it more consistent and easier to analyze.


### Conclusion of Text Preprocessing

We have successfully completed the text preprocessing step, which included tokenization, noise removal, stopword removal, and text normalization through stemming and lemmatization. The data is now clean, consistent, and ready for further analysis.

### Sentiment Analysis

We will begin by performing sentiment analysis to classify the messages as positive, negative, or neutral. For this, we will use the **VADER Sentiment Intensity Analyzer** from the `nltk` library, which is well-suited for analyzing the sentiment of short social messages like SMS texts.


```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to analyze sentiment using VADER
def vader_sentiment(text):
    score = sia.polarity_scores(text)['compound']  # Get the compound score
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply VADER sentiment analysis to the raw 'Message' column (unprocessed)
df['sentiment'] = df['Message'].apply(lambda x: vader_sentiment(str(x)))

# Display the first few rows with the sentiment
df[['Message', 'sentiment']].head()
```

### Interpretation of Sentiment Analysis

After applying VADER sentiment analysis, the messages have been classified as **positive**, **negative**, or **neutral** based on their content:

- The message "Go until jurong point, crazy.. Available only..." has been classified as **positive**, likely due to the positive tone conveyed by words like "crazy" and "available".
- The message "Den only weekdays got special price... Haiz..." is also classified as **positive**, which may reflect the mention of "special price".
- Several messages, such as "Bugis oso near wat..." and "Meet after lunch la..." have been classified as **neutral**, as they likely lack strong positive or negative sentiment.

This analysis helps us understand the general tone of the conversations in the dataset, providing insights into user communication.

### Topic Modeling

Next, we will apply **Latent Dirichlet Allocation (LDA)** to identify the most common topics in the messages. This will give us insight into the main themes discussed in the text.


```python
import gensim
from gensim import corpora

# Prepare the data for LDA
dictionary = corpora.Dictionary(df['normalized_tokens'])
corpus = [dictionary.doc2bow(text) for text in df['normalized_tokens']]

# Train LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)

# Print the topics found by the model
topics = lda_model.print_topics(num_words=5)
for idx, topic in topics:
    print(f"Topic {idx+1}: {topic}")
```

### Interpretation of Topic Modeling

After applying LDA, we have identified five main topics from the SMS messages. Each topic consists of words that frequently co-occur in the messages, giving us insight into common themes:

- **Topic 1**: Includes words like "ok", "sorry", "see", and "come". This topic might represent casual conversations where people are coordinating plans or offering apologies.
- **Topic 2**: Dominated by words like "haha", "okay", "yeah", and "like". This suggests casual, positive conversations, potentially reflecting informal chatting or agreement.
- **Topic 3**: Words such as "hahaha", "please", "class", and "way" suggest humorous or polite conversations, possibly related to group activities like classes or events.
- **Topic 4**: Includes frequent use of "u", "go", "meet", and "got". This topic might represent logistical or planning conversations, likely about meetings or getting together.
- **Topic 5**: Contains words like "lol", "haha", and "im", reflecting humorous or light-hearted interactions, perhaps with expressions of laughter and informal chats.

Each of these topics provides a snapshot of the types of interactions happening within the SMS dataset, with a strong presence of casual and humorous exchanges.

### Word Frequency Analysis

We will now create a frequency distribution of the most commonly used words across the dataset. This will help us understand the overall vocabulary trends in the text messages.


```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Step 1: Join the tokenized words back into a sentence for each row
df['preprocessed_text'] = df['normalized_tokens'].apply(lambda tokens: ' '.join(tokens))

# Step 2: Create the Bag-of-Words model using the preprocessed text
vectorizer = CountVectorizer()  # Initialize CountVectorizer
X = vectorizer.fit_transform(df['preprocessed_text'].dropna())  # Fit and transform the 'preprocessed_text' column

# Step 3: Create a DataFrame of word frequencies
word_freq = pd.DataFrame({'word': vectorizer.get_feature_names_out(), 'frequency': X.sum(axis=0).A1})

# Step 4: Sort the words by frequency
word_freq = word_freq.sort_values(by='frequency', ascending=False)

# Step 5: Display the top 10 most frequent words
print(word_freq.head(10))

# Step 6: Visualize the top 10 most frequent words
plt.figure(figsize=(10, 6))
sns.barplot(x='frequency', y='word', data=word_freq.head(10), palette='viridis')
plt.title('Top 10 Most Frequent Words')
plt.xlabel('Frequency')
plt.ylabel('Word')

# Step 7: Save the plot as a .png file in the specified folder
plot_dir = 'C:/Users/adrco/Projects/nus_sms/plots'
if not os.path.exists(plot_dir):
    os.makedirs(plot_dir)

# Save the plot
plt.savefig(os.path.join(plot_dir, 'word_frequency.png'))

# Show the plot
plt.show()

print(f"Word frequency plot saved at: {os.path.join(plot_dir, 'word_frequency.png')}")

```

### Text Similarity Analysis

In this step, we will measure the similarity between different messages using the **Levenshtein distance**. This metric quantifies how different two strings are by calculating the minimum number of single-character edits (insertions, deletions, or substitutions) needed to transform one string into another. 

The Levenshtein distance is useful in analyzing text similarity because it provides insight into how closely related two messages are, based on their content. A smaller distance indicates higher similarity, while a larger distance reflects more differences between the two texts.

#### Why Levenshtein Distance?
The Levenshtein distance is well-suited for short text messages like SMS because it focuses on character-level changes, capturing the subtle differences between messages that may otherwise seem quite similar or distinct.

#### Example:

Consider the two strings:

- "hello"
- "hallo"

To transform "hello" into "hallo", we only need **one edit** (a substitution of "e" to "a"), so the Levenshtein distance between these two strings is **1**.

In contrast, for the strings:

- "goodbye"
- "hello"

Several edits would be needed (insertions, deletions, and substitutions), so the Levenshtein distance would be **7**.

By applying this to SMS messages, we can identify how similar or different the messages are based on the number of changes required to turn one message into another.


### Data Preparation: Categorizing Message Length and Checking Group Sizes

Before conducting the similarity analysis, we need to prepare the data by categorizing messages based on their length and checking the size of each group.

1. **Convert `length` to Numeric**: 
   - We will convert the `length` column to a numeric format, handling any errors (such as non-numeric values) by coercing them to `NaN`.

2. **Create Message Length Categories**: 
   - Messages will be classified into categories based on their length:
     - **Short**: Messages with fewer than 50 characters.
     - **Long**: Messages with 50 or more characters.
     - **Unknown**: Messages with non-numeric lengths.

3. **Check Group Sizes**: 
   - We will examine the sizes of the following groups:
     - **Country**: The number of messages from each country.
     - **Sentiment**: The number of messages labeled as positive, neutral, or negative.
     - **Message Length Category**: The distribution of messages across the "short", "long", and "unknown" categories.

This preparation step will help us understand the distribution of the data and ensure we handle any irregularities in the message length data.


```python
# Step 1: Convert 'length' to numeric, coercing errors to NaN
df['length_numeric'] = pd.to_numeric(df['length'], errors='coerce')

# Step 2: Create categories based on message length, ignoring NaN values
df['message_length_category'] = df['length_numeric'].apply(lambda x: 'short' if x < 50 else 'long' if pd.notnull(x) else 'unknown')

# Step 3: Check the size of each group by country, sentiment, and message length category
print("### Group Sizes by Country ###")
print(df['country'].value_counts())

print("\n### Group Sizes by Sentiment ###")
print(df['sentiment'].value_counts())

print("\n### Group Sizes by Message Length Category ###")
print(df['message_length_category'].value_counts())
```

### Text Similarity Calculation and Data Preparation

In this step, we prepare the dataset for the **Levenshtein distance** similarity analysis. The main actions include:

1. **Joining Preprocessed Tokens**: 
   - The tokenized words in the `normalized_tokens` column are joined back into full sentences, stored in a new column `preprocessed_text`. This will allow us to compare the text messages using character-level similarity measures.

2. **Defining the Levenshtein Distance Function**: 
   - We define a function to calculate the **Levenshtein distance** between two messages. This metric quantifies the similarity between messages based on the minimum number of single-character edits required to transform one string into another.

3. **Progress Bar for Similarity Calculation**: 
   - A progress bar (using `tqdm`) is added to provide feedback on the progress of the pairwise similarity calculations, which can take a considerable amount of time when dealing with large datasets.
   
4. **Limiting Pairs for Efficiency**: 
   - To avoid excessive computation time, we limit the number of message pairs to a maximum of 5,000 per group.


```python
from nltk.metrics.distance import edit_distance
import itertools
import numpy as np
from tqdm import tqdm  # Import tqdm for progress bar

# Function to calculate similarity (Levenshtein distance) between two text messages
def calculate_similarity(text1, text2):
    return edit_distance(text1, text2)

# Function to calculate average similarity within a group (e.g., by country, sentiment, or message length) with progress bar
def average_similarity_within_group(group, pair_limit=5000):
    pairs = list(itertools.combinations(group['preprocessed_text'], 2))  # Get all combinations of text pairs
    pairs = pairs[:pair_limit]  # Limit to first 'pair_limit' pairs to reduce processing time
    distances = []
    
    # Use tqdm to create a progress bar for the loop
    for pair in tqdm(pairs, desc="Processing Pairs"):
        distances.append(calculate_similarity(pair[0], pair[1]))  # Calculate distance for each pair
    
    return np.mean(distances) if distances else None  # Return the average distance

# Step 1: Join the tokens back into sentences for each message
df['preprocessed_text'] = df['normalized_tokens'].apply(lambda tokens: ' '.join(tokens))
```

### Similarity Analysis by Country

The Levenshtein distance will help us understand how similar the messages from each country are to each other, based on the number of character-level edits required to transform one message into another.

**Key Points**:
- **Group by Country**: Messages are grouped by the country of origin.
- **Limit to 5,000 Pairs**: To reduce computation time, we limit the analysis to the first 5,000 message pairs for each country group.
- **Average Similarity**: The average Levenshtein distance for each country group is calculated, providing insight into the overall similarity of messages within each country.

We will display the first 10 results from the analysis.


```python
# Similarity Analysis by Country (Limit to 5,000 pairs per country group)
print("### Similarity Analysis by Country (First 10 Results) ###")

# Using include_groups=False to avoid the deprecation warning
avg_distances_by_country = df.groupby('country', group_keys=False).apply(
    lambda group: average_similarity_within_group(group, pair_limit=5000), 
    include_groups=False  # Ensure that only the group content is passed to the function
).reset_index(name='avg_similarity')

# Display the first 10 results
print(avg_distances_by_country.head(10))

```

### Part-of-Speech (POS) Tagging

In this step, we will apply **Part-of-Speech (POS) Tagging** to the text messages using **spaCy** (nltk didn't work for unknow reasons), a popular NLP library. POS tagging helps us identify the grammatical structure of the sentences by labeling each word with its part of speech (e.g., noun, verb, adjective).

The goal of this step is to:
- **Analyze common word combinations**: By identifying patterns such as noun-verb pairs or adjective-noun combinations, we can gain insight into the structure and meaning of the messages.
- **Understand message complexity**: POS tagging allows us to assess the complexity of sentence structures used across different messages.



```python
import spacy

# Load the spaCy English model
nlp = spacy.load('en_core_web_sm')

# Function to perform POS tagging using spaCy
def pos_tagging_spacy(text):
    doc = nlp(text)  # Process the text with spaCy
    return [(token.text, token.pos_) for token in doc]  # Return word and its part of speech

# Apply POS tagging to the 'preprocessed_text' column in the DataFrame
df['pos_tags'] = df['preprocessed_text'].apply(pos_tagging_spacy)

# Display the first few rows of the DataFrame with POS tags
print(df[['preprocessed_text', 'pos_tags']].head())
```

### Brief Interpretation of POS Tagging Output

- Proper nouns (`PROPN`), such as "bugi," "oso," and "jurong," are common, indicating place or entity names.
- Verbs (`VERB`), like "go" and "meet," highlight actions in the messages.
- Informal language is captured with interjections (`INTJ`) like "dunno."

This tagging provides a quick overview of word types and the structure of the text, which can be useful for analyzing sentence patterns and complexity.

### Message Complexity by POS Tags

In this step, we will evaluate the complexity of messages by analyzing the diversity of POS (Part-of-Speech) tags in each message. The assumption is that more complex messages will have a greater variety of POS tags (e.g., the inclusion of adjectives, adverbs, and auxiliary verbs alongside nouns and verbs).

The goal of this analysis is to:
- **Understand message complexity**: By quantifying the diversity of POS tags, we can evaluate the syntactic richness of messages.
- **Link complexity to sentiment and message length**: We will investigate whether more complex messages are associated with certain sentiments (positive, negative, neutral) or message lengths (short, long).


```python
from collections import Counter

# Step 1: Function to calculate the number of unique POS tags in a message
def pos_tag_complexity(pos_tags):
    if pos_tags:  # If the message has POS tags
        pos_only = [tag for word, tag in pos_tags]  # Extract only the POS tags
        return len(set(pos_only))  # Return the number of unique POS tags
    else:
        return 0  # Return 0 if no POS tags are available

# Step 2: Apply the function to calculate POS tag complexity for each message
df['pos_complexity'] = df['pos_tags'].apply(pos_tag_complexity)

# Step 3: Display the first few rows to inspect the new 'pos_complexity' column
print(df[['preprocessed_text', 'pos_tags', 'pos_complexity']].head())

# Step 4: Analyze the relationship between POS complexity and sentiment
complexity_by_sentiment = df.groupby('sentiment')['pos_complexity'].mean().reset_index()
print("Average POS Complexity by Sentiment:")
print(complexity_by_sentiment)

# Step 5: Analyze the relationship between POS complexity and message length category
complexity_by_length = df.groupby('message_length_category')['pos_complexity'].mean().reset_index()
print("Average POS Complexity by Message Length Category:")
print(complexity_by_length)
```

1. **POS Complexity by Individual Messages**:
   - The `pos_complexity` values indicate the number of unique POS tags in each message.
   - For example, the first message has a complexity score of 2, meaning it contains only two unique POS tags. In contrast, the third message has a complexity score of 7, reflecting a more diverse grammatical structure.

2. **Average POS Complexity by Sentiment**:
   - **Negative Sentiment**: The average POS complexity is 3.89, indicating that messages classified as negative tend to have a moderate variety of grammatical structures.

   - **Positive Sentiment**: The average POS complexity is 4.02, suggesting that positive messages tend to be more complex than neutral or negative ones, using a wider variety of grammatical structures.

3. **Average POS Complexity by Message Length**:
   - **Long Messages**: With an average POS complexity of 4.91, long messages typically contain more diverse grammatical structures.
   - **Short Messages**: Short messages have an average complexity of 2.45, indicating that they tend to use simpler grammatical constructions.
